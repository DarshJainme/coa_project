# Reason for speedup less than 1 if a single thread

The speedup of less than 1 with a single thread indicates that the parallelized version of the program introduces overhead, which makes it slightly slower than the sequential version when executed with only one thread. Here are a few reasons why this happens:

Parallel Overheads: Even when running with only one thread, a parallel implementation may have additional overhead such as task management, thread creation, and synchronization, which arenâ€™t present in the purely sequential version. These extra tasks take time and can slow down execution.

Setup and Scheduling: In a parallelized version, the program may still set up structures for parallel execution (e.g., dividing tasks, managing workload), even if it ends up running only one thread. This setup adds additional execution time compared to the purely sequential version.

# Why does SIMD increase speedup drastically?

SIMD is a form of data-level parallelism, where the same operation is applied simultaneously across multiple data points. This approach is highly effective for repetitive calculations like those in the Laplacian solver, where each computation is independent and can be performed in parallel.

# Reason for getting a speedup plateau
Physical Cores vs. Threads

Physical cores: These are the actual processing units on a CPU, each capable of executing one task at a time.

Threads: A thread is a path of execution for a process. In a multi-threaded program, multiple threads can be used to execute tasks concurrently.
Most modern CPUs support multi-threading (often called hyper-threading on Intel processors), which allows each physical core to handle more than one thread. For example, a 4-core CPU with hyper-threading might support 8 threads. However, multi-threading does not double the processing power; it only improves resource utilization when there are idle CPU cycles.

True Parallelism vs. Context Switching
When we run a program with threads equal to or fewer than the available physical cores (e.g., 4 threads on a 4-core CPU), each thread can be assigned to its own core. This allows true parallelism, where each thread can run simultaneously without needing to share processing time with other threads.

When the number of threads exceeds the number of physical cores (e.g., 8 threads on a 4-core CPU), the system canâ€™t provide each thread with its own dedicated core. Instead, it needs to time-share each core among multiple threads, a process managed by the operating system.

How Context Switching Works
Context switching is the process by which the operating system quickly switches between threads on the same core. This switching happens so fast that it can give the appearance of simultaneous execution.
However, context switching has an overhead: each time the CPU switches from one thread to another, it needs to save the state of the current thread and load the state of the next one. This takes time and consumes CPU resources, which can slow down performance.
Why Speedup Plateaus
When our program has more threads than cores:

Higher Context-Switching Overhead: With more threads than physical cores, the CPU spends more time switching between threads rather than executing actual work. The added overhead limits speedup because the CPU is frequently interrupted to manage multiple threads on the same core.

Limited CPU Resources per Thread: Physical cores are the real source of computational power. When multiple threads share the same core, they are no longer running truly in parallel; instead, theyâ€™re taking turns, which dilutes the effectiveness of additional threads. Therefore, after a certain number of threads, we get diminishing returns as the performance benefit of adding threads is offset by the lack of available cores.

Diminished Performance Gains: Initially, adding threads to a multi-threaded program increases speedup because it allows better utilization of all available cores. However, once all cores are fully utilized, adding more threads doesnâ€™t increase the computational power. Instead, it just introduces more context switching and contention for core time, leading to a plateau or even a decrease in speedup.

# Why does speedup remains the same on increasing the block size?â€‹

Speedup remains the same after a certain block size because the benefits of tiling saturate while the overheads stabilize as beyond a certain block size, the memory system reaches its bandwidth limit. Even if computations continue to scale, the rate of data transfer cannot increase further, resulting in no additional speedup. When the block size is small, cache efficiency improves significantly. However, once the block size optimally utilizes the cache, increasing it further causes no effect.â€‹

# Why does after a particular block size the speedup decreases?â€‹

When we process larger blocks, more data is required simultaneously. As block size increases, the number of blocks decreases. # Why is speedup for larger matrices greater than smaller matrices?â€‹

Initially, adding threads to a multi-threaded program increases speedup because it allows better utilization of all available cores. However, once all cores are fully utilized, adding more threads doesnâ€™t increase the computational power. Instead, it just introduces more context switching and contention for core time, leading to a plateau or even a decrease in speedup.â€‹

Parallelization overheads (e.g., thread creation, synchronization) are relatively constant, regardless of matrix size.â€‹

In larger matrices, the computational workload dominates the overheads, resulting in better speedup.â€‹

For smaller matrices, the overheads become a significant fraction of the total runtime, reducing the net speedup.â€‹
